```python
import pandas as pd
import numpy as np
#import matplotlib
#import matplotlib.pyplot as plt
import os
import re
```

## Read Folders

```python
df = pd.read_csv('Financial statements_SharpeRatio_RL_modify.csv')
df
```

```python
directory_path = 'before-samemodel-std-edge'
files = os.listdir(directory_path)

all_dfs = []
# 依序讀取每個文件的名稱和内容
for file_name in files:
    # 構件文件的完整路徑
    file_path = os.path.join(directory_path, file_name)
    
    # 檢查是否為文件
    if os.path.isfile(file_path):
        date_match = re.search(r'\d{4}-\d{2}-\d{2}', file_name)
        print(date_match.group())
        
        filtered_df = df[df['date'] == date_match.group()]
        sorted_df = filtered_df.sort_values(by='stock_id')
        sorted_df = sorted_df.reset_index(drop=True)
        
        
        df_edge = pd.read_csv(file_path)
 
        merged_df = pd.merge(sorted_df, df_edge, left_index=True, right_index=True)
        all_dfs.append(merged_df)
        
        print(merged_df.head())
        
final_merged_df = pd.concat(all_dfs)
```

```python
final_merged_df.rename(columns={'stock_id': 'tic', 'sharpe_ratio': 'close'}, inplace=True)
```

```python
final_merged_df.to_csv('Input for Stock-Picked Agent train.csv', index=False)
```
